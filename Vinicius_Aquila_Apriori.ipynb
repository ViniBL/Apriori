## Importando bibliotecas necessárias

```python
from efficient_apriori import apriori
import pandas as pd
```

## Lendo o CSV

```python
df = pd.read_csv("https://raw.githubusercontent.com/ViniBL/Apriori/main/CestaPao.csv")
df.head(27)
```

### Retirando todos os casos em que o item comprado aparece como "NONE"

```python
df = df[df['Item'] != "NONE"]
df.to_csv("CestaPaoOtimizado.csv")
```

Essa função irá ser a responsável por criar uma lista extraindo os dados do csv "CestaPaoOtimizado", basicamente, ela adicionará ao "transactions" os elementos feitos pela mesma transação

```python
def load_dataset(arquivo):
    with open(arquivo, 'r') as f:
        content = f.readlines()[1:]
    transactions = []
    prev_tid = -1
    for t in content:
        t = t.strip().split(',')[-2:]
        tid = t[0]
        item = t[1]
        if prev_tid != tid:
            prev_tid = tid
            transactions.append([item])
        else:
            transactions[-1].append(item)
    return transactions

```

Agora "lista" será uma lista de listas, onde cada lista se refere a uma transição com seus respectivos itens

```python
lista = load_dataset('CestaPaoOtimizado.csv')
```

```python
lista
```

## Executando o Apriori e mostrando o resultado

O apriori é executado com o suporte mínimo como 0.02 e a confiança mínima como 0.2 

```python
itemsets, rules = apriori(lista, min_support=0.02, min_confidence = 0.2)
```

```python
itemsets
```

Mostrando as métricas de cada regra

```python
print("Confiança: conf, Suporte: supp, Lift: lift, Convicção: Conv\n")
for i in rules:
    print("Regra: ",i)
    print("================================================================================================\n")
```
